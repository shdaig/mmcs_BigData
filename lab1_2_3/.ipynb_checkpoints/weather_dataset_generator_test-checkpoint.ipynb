```python
N = 48 * 365
```

```python
init_datetime = datetime.fromisoformat('2022-01-01T00:00:00')
for i in range(N):
    init_datetime = init_datetime + timedelta(minutes=30)
print(init_datetime)
```

```python
import random

a = -0.8
b = 0.8

temp_min = -8
temp_max = 20

x_year = np.linspace(-np.pi / 2, 3 * np.pi / 2, N)

randomizer = np.random.rand(N // 48) * (b - a) + a

y_year = np.sin(x_year - 0.2) * ((temp_max - temp_min) / 2) + (((temp_max - temp_min) / 2) + temp_min)

for i in range(365):
    y_year[i * 48 : (i + 1) * 48] += randomizer[i]

# plt.plot(y_year[9312 - (48 * 5) : 9312 + (48 * 5)])
plt.plot(y_year)
plt.show()
print(np.argmax(y_year))
```

```python
x_day = np.linspace(-np.pi / 2, 3 * np.pi / 2, 48)
y_day = np.sin(x_day - np.pi / 3.5)
# y_day = np.sin(x_day)

plt.plot(y_day)
```

```python
for i in range(365):
    y_year[48 * i : 48 * (i + 1)] += y_day
    
for i in range(1, 364):
    for j in range(-3, 3):
        y_year[48 * i + j] = np.mean(y_year[48 * i + j - 3 : 48 * i + j + 3])

a = -0.1
b = 0.1    
    
randomizer = np.random.rand(N) * (b - a) + a

y_year += randomizer
```

```python
y_year = np.round(y_year, decimals=1)
```

```python
d1 = 0
d2 = 365
# print(y_year[48 * d1 : 48 * d2])
print(y_year.shape)
plt.plot(y_year[48 * d1 : 48 * d2])
plt.show()
```

```python
k = 0.28
```

```python
temp_mean = (temp_max + temp_min) / 2 
temp_mean
```

```python
y_pressure = 760 + (temp_mean - y_year) * k
y_pressure = np.round(y_pressure, decimals=1)
```

```python
plt.plot(y_pressure)
```

```python
max_humidity = 86
min_humidity = 61
avg_humidity = 72
hk = (max_humidity - min_humidity) / (np.max(y_year) - np.min(y_year))
```

```python
y_humidity = avg_humidity + (temp_mean - y_year) * hk
y_humidity = np.round(y_humidity, decimals=0)
```

```python
plt.plot(y_humidity)
```

```python
x_sunshine = np.linspace(-np.pi / 2, 3 * np.pi / 2, 365)

sunshine_min = 7
sunshine_max = 17.5

y_sunshine_daily = np.sin(x_sunshine + (np.pi / 365) * 19) * ((sunshine_max - sunshine_min) / 2) + (((sunshine_max - sunshine_min) / 2) + sunshine_min)
# y_day = np.sin(x_day)
# print(np.argmin(y_sunshine))

y_sunshine_daily = np.round(y_sunshine_daily, decimals=2)

plt.plot(y_sunshine_daily)
```

```python
print(y_sunshine_daily[79])
print(y_sunshine_daily[172])
print(y_sunshine_daily[266])
print(y_sunshine_daily[355])
```

```python
y_sunshine = np.full((N, ), 1.)
for i in range(365):
    y_sunshine[i * 48 : (i + 1) * 48] *= y_sunshine_daily[i]
    
plt.plot(y_sunshine)
```

```python
init_datetime = datetime.fromisoformat('2022-01-01T00:00:00')
time_arr = []
for i in range(N - 1):
    init_datetime = init_datetime + timedelta(minutes=30)
    time_arr.append(init_datetime)
    
time_arr = np.array(time_arr)
```

```python
weather_df = pd.DataFrame({'date': time_arr, 'temperature': y_year[:N-1], 'pressure': y_pressure[:N-1], 'humidity': y_humidity[:N-1], 'sunshine': y_sunshine[:N-1]})
```
