```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, timedelta 

import time
import random
```

```python
N = 48 * 365
```

**Генерация температуры**

```python
temp_min = -22
temp_max = 18

temp_var_min = -0.8
temp_var_max = 0.8
```

```python
x_temperature = np.linspace(-np.pi / 2, 3 * np.pi / 2, N)

y_temperature = np.sin(x_temperature - 0.2) * ((temp_max - temp_min) / 2) + (((temp_max - temp_min) / 2) + temp_min)

plt.plot(y_temperature)
```

```python
randomizer = np.random.rand(N // 48) * (temp_var_max - temp_var_min) + temp_var_min

for i in range(365):
    y_temperature[i * 48 : (i + 1) * 48] += randomizer[i]

plt.plot(y_temperature)
```

```python
x_day = np.linspace(-np.pi / 2, 3 * np.pi / 2, 48)
y_day = np.sin(x_day - np.pi / 3.5)

plt.plot(y_day)
```

```python
for i in range(365):
    y_temperature[48 * i : 48 * (i + 1)] += y_day

d1 = 0
d2 = 5
plt.plot(y_temperature[48 * d1 : 48 * d2])
```

```python
for i in range(1, 364):
    for j in range(-3, 3):
        y_temperature[48 * i + j] = np.mean(y_temperature[48 * i + j - 3 : 48 * i + j + 3])

a = -0.1
b = 0.1    
    
randomizer = np.random.rand(N) * (b - a) + a

y_temperature += randomizer
```

```python
y_temperature = np.round(y_temperature, decimals=1)
```

```python
d1 = 0
d2 = 10 # 365 - весь год, толстая синусоида
plt.plot(y_temperature[48 * d1 : 48 * d2])
plt.show()
```

**Генерация давления**

```python
k = 0.28
```

```python
temp_mean = (temp_max + temp_min) / 2 
temp_mean
```

```python
y_pressure = 760 + (temp_mean - y_temperature) * k
y_pressure = np.round(y_pressure, decimals=1)
```

```python
plt.plot(y_pressure)
```

**Генерация влажности**

```python
max_humidity = 86
min_humidity = 61
avg_humidity = 72
hk = (max_humidity - min_humidity) / (np.max(y_temperature) - np.min(y_temperature)) 
```

```python
y_humidity = avg_humidity + (temp_mean - y_temperature) * hk
y_humidity = np.round(y_humidity, decimals=0)
```

```python
plt.plot(y_humidity)
```

**Генерация длины дня**

```python
x_sunshine = np.linspace(-np.pi / 2, 3 * np.pi / 2, 365)

sunshine_min = 7
sunshine_max = 17.5

y_sunshine_daily = np.sin(x_sunshine + (np.pi / 365) * 19) * ((sunshine_max - sunshine_min) / 2) + (((sunshine_max - sunshine_min) / 2) + sunshine_min)


y_sunshine_daily = np.round(y_sunshine_daily, decimals=2)

plt.plot(y_sunshine_daily)
```

```python
print(y_sunshine_daily[79]) # 20 марта
print(y_sunshine_daily[172]) # 21 июня
print(y_sunshine_daily[266]) # 23 сентября
print(y_sunshine_daily[355]) # 21 декабря
```

```python
y_sunshine = np.full((N, ), 1.)
for i in range(365):
    y_sunshine[i * 48 : (i + 1) * 48] *= y_sunshine_daily[i]
    
plt.plot(y_sunshine)
```

**Визуализация**

```python
fig = plt.figure();
ax = fig.add_subplot(2, 2, 1)
ax.plot(y_temperature)
ax = fig.add_subplot(2, 2, 2)
ax.plot(y_pressure)
ax = fig.add_subplot(2, 2, 3)
ax.plot(y_humidity)
ax = fig.add_subplot(2, 2, 4)
ax.plot(y_sunshine)
```

**Генерация датафрейма**

```python
init_datetime = datetime.fromisoformat('2022-01-01T00:00:00')
time_arr = []
for i in range(N - 1):
    init_datetime = init_datetime + timedelta(minutes=30)
    time_arr.append(init_datetime)
    
time_arr = np.array(time_arr)
```

```python
weather_df = pd.DataFrame({'date': time_arr, 'temperature': y_temperature[:N-1], 'pressure': y_pressure[:N-1], 'humidity': y_humidity[:N-1], 'sunshine': y_sunshine[:N-1]})
```
